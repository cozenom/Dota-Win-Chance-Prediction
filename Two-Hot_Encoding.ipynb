```python
def encode(data):
    res = []
    for i in range(0,len(data)):
        t0 = data.iloc[i][0]
        t1 = data.iloc[i][1]
        tdict0 = {x: 1 for x in t0}
        tdict0.update({y+135: 1 for y in t1})
        res += [tdict0]
    return pd.DataFrame.from_dict(res)#.fillna(value=0)
```
