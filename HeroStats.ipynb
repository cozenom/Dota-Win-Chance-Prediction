```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# https://docs.opendota.com/#tag/hero-stats%2Fpaths%2F~1heroStats%2Fget
heroStats = 'https://api.opendota.com/api/heroStats'
r = requests.get(heroStats)
```

```python
data = json.loads(r.text)
df = pd.DataFrame(data)
#df.head(10)
```

```python
print(df.columns)
drop = ['primary_attr', 'attack_type', 'roles', 'img', 'icon', 'base_health', 'base_health_regen', 'base_mana',
       'base_mana_regen', 'base_armor', 'base_mr', 'base_attack_min',
       'base_attack_max', 'base_str', 'base_agi', 'base_int', 'str_gain',
       'agi_gain', 'int_gain', 'attack_range', 'projectile_speed',
       'attack_rate', 'move_speed', 'turn_rate', 'cm_enabled', 'legs',
        'turbo_picks', 'turbo_wins',
        'pro_ban', 'pro_win', 'pro_pick',  
        'null_pick', 'null_win']
```

```python
df.drop(drop, axis=1, inplace = True)
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
df.sort_values(by=['localized_name'])
```

```python
df.sort_values(by=['id'])
```

```python
# Create aggregate pick/win columns
df['Agg_pick']=df['1_pick'] + df['2_pick'] + df['3_pick'] + df['4_pick'] + df['5_pick'] + df['6_pick'] + df['7_pick'] + df['8_pick']
df['Agg_win']=df['1_win'] + df['2_win'] + df['3_win'] + df['4_win'] + df['5_win'] + df['6_win'] + df['7_win'] + df['8_win']
df['Agg_win_rate']=df['Agg_win']/df['Agg_pick']

df['Agg_pick_rate']=df['Agg_pick']/df['Agg_pick'].sum()

df.sort_values(by=['Agg_win_rate'], ascending=False)
df2 = df[['id','localized_name','Agg_win_rate']]
df2
```

```python
df3 = df.sort_values(by=['Agg_pick_rate'], ascending=False)
df3 = df3[['id','localized_name','Agg_pick_rate']]
df3
```

```python
ax = df3.plot.bar(x='id', y='Agg_pick_rate', rot=0, legend=False, figsize=(20, 10), color="green")
ax.set(title="Aggregate pick rate distribution",xlabel="", ylabel="")
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
ax
```
